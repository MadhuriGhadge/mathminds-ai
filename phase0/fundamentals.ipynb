#### List vs. Tuple
##### Python lists and tuples are sequence data types that store ordered collections of items. While lists are mutable and ideal for dynamic, homogeneous data, tuples are immutable, making them suitable for fixed, heterogeneous data.

```python
#define a list
colors = ['red','green','blue','yellow']
print(colors)

# define a tuple
person = ("Madhuri Ghadge",21,"python programmer","Maharashtra")
print(person)



"""
a list doesn’t have a fixed length since it’s mutable. Therefore, it’s natural to use homogeneous elements to have some structure in the list.
A tuple, on the other hand, has a fixed length so the position of elements can have meaning, supporting heterogeneous data. """
```

```python
#define a list object using a literal. 
#A list literal is a comma-separated sequence of objects enclosed in square brackets
list = ['item_0', 'item_1', ..., 'item_n']
# -> lists are ordered
# -> hoterogeneous
```

```python
def fibonacci_generator(stop):
    current_fib, next_fib = 0, 1
    for _ in range(0, stop):
       fib_number = current_fib
       current_fib, next_fib = next_fib, current_fib + next_fib
       yield fib_number

fibonacci_generator(10)
```

```python
list(fibonacci_generator(10))
```

```python
[*fibonacci_generator(10)]
```

#### Dictionaries

```python

student = {
    "name":"madhuri",
    "age":21,
    "CGPA":7.81
}

print(student['name'])

#add/update
student["grade"] = "A"
print(student)

# delete
del student['CGPA']
print(student)

print(student.keys())
print(student.values())
print(student.items())
```

```python
movie = {
    "title":"Jaudya na balasaheb",
    "year" : 2019,
    "rating" : 5,
    "genre" : "romcom"
}

print(movie.keys())
print(movie.values())
movie["rating"] = 8
del movie['genre']
print(movie)
```

```python
students = [
    {"name":"Aarav","marks":88},
    {"name":"Madhuri","marks":88.33},
    {"name":"Akshada","marks":88}
]

for student in students:
    print(student["name"],"->",student["marks"])
```

```python
books = [
    {"title":"Shreeman Yogi","author":"Ranjit Desai","price":900},
    {"title":"Mrutunjay","author":"Shivaji Sawant","price":500},
    {"title":"Chhava","author":"Shivaji Sawant","price":590}
]

#Print the most expensive book’s title

most_expensive = max(books, key=lambda x:x['price'])
print(most_expensive)

total_price = sum(book['price'] for book in books)
print(total_price)

```

#### conditionals

```python
## simple if


n = int(input("enter a number:"))
if n%2 == 0:
    print("number is even")
```

```python
# if .... else

n = int(input("enter a number:"))
if n%2 == 0:
    print("number is even")
else:
    print("number is odd")
```

```python
# if ........ elif ...... else
n = int(input("enter a number:"))
if n == 0:
    print("number is zero")
elif n%2 == 0:
    print("number is even")
else:
    print("number is odd")
```

```python
# nested if

n = int(input("enter a number:"))
if n%3 == 0:
    if n% 5 == 0:
        print("number is divisible by 3 and 5")
else:
    print("number is not divisible by 3")
```
